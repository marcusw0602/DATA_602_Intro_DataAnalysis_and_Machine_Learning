# Predicting postive and negative cardiovascular disease cases

# Abstract 
Cardiovascular disease or known as CVD is a group of conditions that involve the body's heart and blood vessels. Currentlu CVD is the leading cause of death in the United State where it is reported that one person dies every 37 second from CVD. CVD has many facets to it, meaning it cause many complications such as chest pain, stroke, and even heart attacks. It is stated by 2030 more than over one million americans would die of CVD and its complications every year after. The most important behavioural risk factors of heart disease and stroke are unhealthy diet, physical inactivity, tobacco use and harmful use of alcohol. The effects of behavioural risk factors may show up in individuals as raised blood pressure, raised blood glucose, raised blood lipids, and overweight and obesity. These “intermediate risks factors” can be measured in primary care facilities and indicate an increased risk of developing a heart attack, stroke, heart failure and other complications.

# Machine learning models

### Features
- AGE:  integer (years of age)
- WEIGHT: integer (kg)
- GENDER: categorical (0: female, 1: male)
- AP_HIGH: systolic blood pressure, integer
- AP_LOW: diastolic blood pressure, integer 
- CHOLESTEROL: categorical (1: normal, 2: above normal, 3: well above normal)
- GLUCOSE: categorical (1: normal, 2: above normal, 3: well above normal)
- BMI: integer

### Target
- CARDIO_DISEASE:  categorical (0: no, 1: yes)

### Logistic Regression
We first are going to start with the logistic regression model to predict the positive and negative cases of cardiovascular disease using the features below.
What you see below is the features of and target being seperated into two values 'features' and 'target', I then take those fields and further seperate their values in two training and two testing sets. For the testing set I am only going to use 20% of the original trained set. I am also standardizing the X trained and test set for regularization.

```python
from sklearn.model_selection import train_test_split
features=['age','gender','weight','ap_hi','ap_lo','cholesterol','gluc']
target=['cardio']
X=df[features].values
y=df[target].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state=42)
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

### Cross validation of a Logistic Regression with Regularization
To gain the best overall model score I am starting a cross validation process using grid search cross validation. In doing this I hope to gain the best insight into which parameters to use to strenghthen my logistic model. 
Below I am creating a grid of my C and penalty values that I would like the grid search to run through to determine which parameters will give me the best model score. 

```python
grid = {'C': [0.1, 1, 10, 100, 1000], 'penalty' :['l1', 'l2']}
```

Here I am defining my logistic regression model to have a max iteration of 1000 and the solver to be 'saga'. I am then creating the grid search function to include my logistic model, grid parameters, and cross validation of. 

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
log = LogisticRegression(max_iter= 10000, solver = 'saga')
clf = GridSearchCV(estimator = log, param_grid= grid, n_jobs = -1, verbose= 2, cv=5, return_train_score=True)
clf.fit(X_train, y_train)
```

The following frame below is the results from creating and fitting my X-train and y-train sets. What you can also see is that the models are list from left to right with the best model starting on the left side and weaker down the rigth. From looking at model 8 and 6 you can see that the split test and trained score are nearly identical with the exception of model 8's split2_test_score being higher than model 6. Therefore we can see that model 8 with the parameter of "C:1000 and Penalty:L1" is the model to use when trying to predict the accuracy.

```python
pd.DataFrame(clf.cv_results_).sort_values('mean_test_score', ascending=False).T
```

```python
print("best parameters  are: ", clf.best_params_)
```

### Applying best model and establishing the cross validation test score
Below I am now applying the best model found from the grid search cross validation, then I am establishing the cross validation test score by finding the mean and standard deviation which you can see below is 71% 

```python
from sklearn.model_selection import cross_validate
log = clf.best_estimator_
cv_fivefold = cross_validate(estimator = log, 
                             X=X_train,
                             y=y_train,
                             cv =5, 
                             return_train_score= True,
                             return_estimator= True )
```

```python
test_mean = cv_fivefold['test_score'].mean()
test_std = cv_fivefold['test_score'].std()
print('The best regularized test score is: %.3f  +/- %.3f'%(test_mean, test_std))
```

### Logistic Regression Confusion Matrix
Below I am graphing a matrix that will display the trained values to our test values to determine how accurate our logistic regression model is in determining positive and negative cases of cardiovascular disease.

```python
from sklearn import metrics
from sklearn.metrics import confusion_matrix, classification_report
log.fit(X_train, y_train)
preds = log.predict(X_train)
preds_test = log.predict(X_test)
cm = metrics.confusion_matrix(y_train, preds)
cm_test = metrics.confusion_matrix(y_test, preds_test)
```

```python
score = log.score(X_train, y_train)
print('We can see here that the trained score for this model is:',score)
test_score = log.score(X_test, y_test)
print('We can see here that the test score for this model is:',test_score)
```

We can see the below model graph did report a 71% match between our testing and training sets. The prediction for trained negative cases matched 18,676 times with a fail of 5,892 cases being positive. As well as the positive predictions match 16,372 times and a fail of 8,034 cases being negetive.

```python
plt.figure(figsize=(9,9))
sns.heatmap(cm, annot=True, fmt=".3f", linewidths=.5, square = True)
plt.ylabel('True Values');
plt.xlabel('Predicted Values');
all_sample_title = 'Accuracy Score for the training dataset: {0}'.format(score)
plt.title(all_sample_title, size = 15);
plt.show()
```

```python
print(classification_report(y_train, preds))
```

Now when looking at the test prediction below we can see that the accuracy is similar to that of the trained sets. With predict negative values matching actual values 7,876 times along with positive predictions matching 7,183 times. 

```python
plt.figure(figsize=(9,9))
sns.heatmap(cm_test, annot=True, fmt=".3f", linewidths=.5, square = True);
plt.ylabel('True values');
plt.xlabel('Predicted Values');
all_sample_title = 'Accuracy Score for test dataset: {0}'.format(test_score)
plt.title(all_sample_title, size = 15);
plt.show()
```

```python
print(classification_report(y_test, preds_test))
```

### Decision Tress Classifier
With having one other model creating predictions on cardiovascular cases I wanted to use decision tress to try and increase the accuracy of my predictions and strengthen my model. You can see below I am declaring the tree classifier to have a max leaf node of 100 with using the criterion being 'entropy' and setting my max depth to 10. In doing this I am trying to yield the best accuracy I can for predictions of positive and negative cases. In doing this I am able to achieve a score of 73% for both trained and test values, this being a 2% increase over the previous logistic model.

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(max_leaf_nodes= 100,criterion= 'entropy', max_depth= 10)
cv_fivefold1 = cross_validate(estimator = dtc, 
                              X=X_train, 
                              y=y_train, 
                              cv =5, 
                              return_train_score= True, 
                              return_estimator= True )
```

```python
dtc.fit(X_train, y_train)
score_dtc = dtc.score(X_train,y_train)
test_mean = cv_fivefold1['test_score'].mean()
test_std = cv_fivefold1['test_score'].std()
print('best decision tree test score: %.3f  +/- %.3f'%(test_mean, test_std))
print('The is the trained score using the desicion tree classifier: ',dtc.score(X_train,y_train))
score_dtc_test = dtc.score(X_test,y_test)
print('The is the test score using the desicion tree classifier: ',dtc.score(X_test,y_test))
```

### Decision Tree Confusion Matrix

```python
dt_preds = dtc.predict(X_train)
dt_preds_test = dtc.predict(X_test)
from sklearn import metrics
cm_dt = metrics.confusion_matrix(y_train, dt_preds)
cm_dt_test = metrics.confusion_matrix(y_test, dt_preds_test)
```

With applying the same matrix to the our decision matrix we can see that there were an increase of 543 predicted case that matches our actual values for negative cases. As well as an increase of 568 predicted case for positive reported cases. The same can be said for the tes

```python
plt.figure(figsize=(9,9))
sns.heatmap(cm_dt, annot=True, fmt=".3f", linewidths=.5, square = True)
plt.ylabel('True Values');
plt.xlabel('Predicted Values');
all_sample_title = 'Accuracy Score for training data in Decision Trees: {0}'.format(score_dtc)
plt.title(all_sample_title, size = 15);
plt.show()
```

```python
print(classification_report(y_train, dt_preds))
```

```python
plt.figure(figsize=(9,9))
sns.heatmap(cm_dt_test, annot=True, fmt=".3f", linewidths=.5, square = True)
plt.ylabel('True Values');
plt.xlabel('Predicted Values');
all_sample_title = 'Accuracy Score for testing data in Decision Trees: {0}'.format(score_dtc_test)
plt.title(all_sample_title, size = 15);
plt.show()
```

```python
print(classification_report(y_test, dt_preds_test))
```
